# Feature selection, elección y cálculo de métricas para validar los modelos

```python
#Previo a utilizar las tecnicas de reducción de dimensionalidad se procede a transformar las variables categoricas
#a variables continuas.
df=pd.get_dummies(df,drop_first=True)
```

```python
#Se selecciona las variables del dataset para los ejes x e y.
X = df.drop("Weight", axis=1)
y = df["Weight"]

df
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y)
```

```python
#Se selecciona el modelo de Regresión Lineal para evaluar sus métricas y performance a partir de los datos
#previamente trabajados.
from sklearn.linear_model import  LinearRegression as LR
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#Utilizamos Sequential Feature Selector para seleccionar las 9 variables más representativas.
#A su vez se genera un proceso por default de cross validation utilizando 5 folds sobre el
#modelo de regresión lineal.
from mlxtend.feature_selection import SequentialFeatureSelector as SFS

sfs = SFS(LR(),
          k_features=9)
```

```python
#Se entrena al Feature Selector
sfs.fit(X_train, y_train)
```

```python
#Se obtienen las 9 features más representativas
sfs.k_feature_names_
```

```python
#A partir de las features anteriores se pone a entrenar y luego testear el modelo
#de Regresión Lineal para obtener como resultado su performance.
X = df.loc[:,['Height',
 'FAF',
 'CAEC_Sometimes',
 'NObeyesdad_Normal_Weight',
 'NObeyesdad_Obesity_Type_I',
 'NObeyesdad_Obesity_Type_II',
 'NObeyesdad_Obesity_Type_III',
 'NObeyesdad_Overweight_Level_I',
 'NObeyesdad_Overweight_Level_II']]
y = df["Weight"]

X_train,X_test,y_train,y_test = train_test_split(X,y)

model=LR()
model.fit(X_train,y_train)
y_pred_test=model.predict(X_test)
r2_score(y_pred_test,y_test)
```
